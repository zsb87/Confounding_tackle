```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 15}

matplotlib.rc('font', **font)

fig, ax = plt.subplots(figsize = (9,6))

n_groups = 6


LR = (0.474,0.5279, 0.549, 0.484, 0.438, 0.5)
RF = (0.357,0.5808, 0.5852, 0.5079, 0.6204, 0.5)
SVM = (0.05,0.5055, 0.333, 0.326, 0.35, .25)


index = np.arange(n_groups)
bar_width = 0.2

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index - 0.5*bar_width, LR, bar_width,
                alpha=opacity, color='b',
                label='kmeans(2)+LR')

rects2 = ax.bar(index+0.5*bar_width, RF, bar_width,
                alpha=opacity, color='g',
                label='kmeans(2)+RF')

rects3 = ax.bar(index + 1.5*bar_width, SVM, bar_width,
                alpha=opacity, color='y',
                label='kmeans(2)+SVM')


# ax.set_xlabel('Group')
ax.set_ylim(0,0.8)
ax.set_ylabel('Positive Class F-score')
# ax.set_title('Positive Class F-score with Different Class Balancing Technique')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('Baseline','Random', 'SMOTE', 'SMOTE(SVM)', 'SMOTE(b1)', 'SMOTE(b2)'))
ax.legend()

fig.tight_layout()
plt.show()
```
